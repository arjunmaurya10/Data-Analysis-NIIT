```python
def hist_box(*s):
    for col in s:
        fig,ax=plt.subplots(1,2,figsize=(15,4))
        ax=ax.ravel()
        sns.histplot(x=col,bins=10,ax=ax[0],data=wdf)
        ax[0].set_title(f"distribution of {col} using histogram")
        ax[0].set_xlabel(f"{col}")
        ax[0].set_ylabel(f"frequency of {col}")
        
        
        sns.boxplot(y=wdf[col],ax=ax[1])
        ax[1].set_title(f"distribution using boxplot of {col}")
```

**PCA**

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pca.fit(scale_df)
features=range(pca.n_components_) #no. of principle components  
transformed_data=pca.transform(scale_df)
```

```python
plt.bar(features,pca.explained_variance_ratio_)
plt.xlabel('feature')
plt.ylabel('variance')
plt.xticks(features)
```

```python
#axis (0,1,2,3)
pca2=PCA(n_components=2)
transformed_data2=pca2.fit_transform(scale_df)
transformed_data2.shape
```

```python
plt.scatter(transformed_data2[:,0],transformed_data2[:,1])
plt.xlabel('pca1')
plt.ylabel('pca2')
plt.title('Dimension Reduced')
```

# KMeans

**with dimension reduced**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score,silhouette_samples
```

```python
distortion=[]
for i in range(2,10):
    model=KMeans(n_clusters=i).fit(transformed_data2)
    distortion.append(model.inertia_) # how much distance (WSS)
    
plt.figure(figsize=(10,7))
plt.grid(True)
plt.plot(range(2,10),distortion)
plt.ylabel('Distortion')
plt.xlabel('K value')
plt.title("Elbow Curve")
```

```python
for i in range(2,10):
    model=KMeans(n_clusters=i).fit(transformed_data2)
    model.predict(transformed_data2)
    score=silhouette_score(transformed_data2,model.labels_,metric="euclidean")
    print("k:",i,"score",score)
```

```python
kmodel = KMeans(n_clusters=3).fit(transformed_data2)
```

```python
print(kmodel.labels_)
```

```python
plt.scatter(transformed_data2[:,0],transformed_data2[:,1],c=kmodel.labels_,cmap='rainbow')
plt.xlabel('pca1')
plt.ylabel('pca2')
plt.title('KMeans with reduced dimension')
```

# DBSCAN

**With PCA**

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
```

```python
neighbours=NearestNeighbors(n_neighbors=5)
neigh_fit = neighbours.fit(transformed_data2)
distance, index=neigh_fit.kneighbors(transformed_data2)
```

```python
distance = np.sort(distance,axis=0)
distances=distance[:,1]
plt.plot(distances)
plt.axhline(y=0.4,color='r')
plt.grid()
#plt.yticks(np.arange(0,130,10))
plt.ylim()
```

```python
eps=0.4
for i in range(5,20):
    db=DBSCAN(eps=eps,min_samples=i)
    db.fit(transformed_data2)
    labels=db.labels_
    
    new_labels=labels+2
    y=np.bincount(new_labels)
    z=np.nonzero(y)[0]
    cluster=list(zip(z,y[z]))
    print("min_samples:",i,"cluster:",cluster)
```

```python
for i in range(2,15):
    dmodel=DBSCAN().fit(transformed_data2)
    dmodel.fit(transformed_data2)
    score=silhouette_score(transformed_data2,dmodel.labels_,metric="euclidean")
    print("k:",i,"score",score)
```

```python
db=DBSCAN(eps=eps,min_samples=8)
db.fit(transformed_data2)
labels=db.labels_
```

```python
plt.scatter(transformed_data2[:,0],transformed_data2[:,1],c=labels,cmap='rainbow')
plt.xlabel('pca1')
plt.ylabel('pca2')
plt.title('DBSCAN with reduced dimension')
```
