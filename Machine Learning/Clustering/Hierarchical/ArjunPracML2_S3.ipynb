```python
cardf['Model'].value_counts()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
cat_col = ['Model','Vehicle_type','Manufacturer']
cardf[cat_col] = cardf[cat_col].astype('category')
```

```python
label_encoder =preprocessing.LabelEncoder()

for col in cat_col:
    cardf[col] = label_encoder.fit_transform(cardf[col])
```

```python
cardf
```

```python
cardf['Engine_size'].value_counts()
```

```python
scale_df=pd.DataFrame(StandardScaler().fit_transform(cardf),columns=cardf.columns)
```

```python
scale_df
```

**creating data for Dendrogram** 

```python
x=scale_df.iloc[:,[2,4]].values
```

```python
x
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

**ward linkeage**

```python
# dendogram
plt.figure(figsize=(30,10))
dendrogram = dendrogram(linkage(x,method='ward',metric='euclidean'),labels=scale_df.index)
plt.axhline(y=7.5,color='r',linestyle ='--')
ax=plt.gca()
ax.tick_params(axis='x',which='major',labelsize=10)
plt.title('dendrogram',fontsize=30)
plt.xlabel('Sales',fontsize=20)
plt.ylabel('Euclidean distance',fontsize=20)
```

**ModelAGGLO ward**

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
```

```python
y_hc = hc.fit_predict(x)
```

```python
y_hc
```

```python
plt.scatter(x[y_hc == 0,0],x[y_hc ==0,1],s=50,c='red',label='good sale ')
plt.scatter(x[y_hc == 1,0],x[y_hc ==1,1],s=50,c='blue',label='no sale')
plt.scatter(x[y_hc == 2,0],x[y_hc ==2,1],s=50,c='green',label='sale and price moderate')
plt.scatter(x[y_hc == 3,0],x[y_hc ==3,1],s=50,c='cyan',label='high price no sale')
#plt.scatter(x[y_hc == 4,0],x[y_hc ==4,1],s=50,c='magenta',label='sensible')
plt.title(' Profit by Sales')
plt.ylabel('Price in thousands')
plt.xlabel('Sales in thousands')
plt.legend()
plt.show()
```

**Single linkage**

```python
# dendogram
plt.figure(figsize=(30,10))
dendrogram = dendrogram(linkage(x,method='single',metric='euclidean'),labels=scale_df.index)
plt.axhline(y=2.0,color='r',linestyle ='--')
ax=plt.gca()
ax.tick_params(axis='x',which='major',labelsize=10)
plt.title('dendogram',fontsize=30)
plt.xlabel('Sales',fontsize=20)
plt.ylabel('Euclidean distance',fontsize=20)
```
