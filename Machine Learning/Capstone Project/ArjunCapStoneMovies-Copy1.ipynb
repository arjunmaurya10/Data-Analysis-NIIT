```python
# created a function to convert json to str(string) format
def json_to_string(row):
    json_data = json.loads(row)
    values=[data.get("name") for data in json_data]   
    return values[:3] 
```

```python
#extracting the genres 
g_genre = maindf["genres"].apply(json_to_string)
maindf['genre1']=g_genre.apply(lambda x:x[0] if len(x)>=1 else None)
```

```python
#extracting Directors
maindf['crew'] = maindf['crew'].apply(json.loads)
ex_dir = lambda x: [per['name'] for per in x if per['job']=='Director']
dir_name = maindf['crew'].apply(ex_dir)
maindf['Directors'] = dir_name.apply(lambda x:x[0] if len(x)>=1 else None)
```

```python
# extracting lead actor from cast
lead_n = maindf['cast'].apply(json_to_string)
maindf['Lead_actr']=lead_n.apply(lambda x:x[0] if len(x)>=1 else None)
```

```python
# created a function to convert json to str(string) format
def jts(row):
    json_data = json.loads(row)
    values=[data.get("character") for data in json_data]   
    return values[:3] #or else[:1]
```

```python
#extracting lead char 
lead_c = maindf['cast'].apply(jts)
maindf['Lead_Char']=lead_c.apply(lambda x:x[0] if len(x)>=1 else None)
```

```python
actor_counts = []

# Iterate over each row in the DataFrame
for index, row in maindf.iterrows():
    cast_data = json.loads(row['cast'])
    actor_count = len(cast_data)
    actor_counts.append(actor_count)

# Add the actor_counts as a new column in the DataFrame
maindf['actor_count'] = actor_counts

# Print the DataFrame with the actor_count column
print(maindf[['title', 'actor_count']])
```

```python
mean_vote = maindf['vote_average'].mean()
```

```python

rating = []
for i in maindf['vote_average']:
    if (i > mean_vote) :
        rating.append('Good')
    elif ((i >= (mean_vote/2)) & (i <= mean_vote)) :
        rating.append('Average')
    elif (i < (mean_vote/2)):
        rating.append('Poor')
maindf['Quality'] = rating

```
