```python
import re
import nltk
import spacy
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nlp=spacy.load('en_core_web_lg')
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
mer_df['plot'][0]
```

```python
def clean_n_normal(text): # func to clean and normalize the text in plot
    text = text.lower() # cnvert to lower
    text = re.sub(r'[^a-zA-Z\s]', '', text) #remove unwanted char and digit
    tokens = word_tokenize(text) # tokenize 
    stop_words = set(stopwords.words('english')) # removng stopwords
    fil_tk = [token for token in tokens if token not in stop_words]
    normal_txt = ' '.join(fil_tk) # joing back to form normalze txt
    
    doc = nlp(normal_txt) #lemma with spacy
    norm_tk = [token.lemma_ for token in doc]
    normal_txt = ' '.join(norm_tk) #joing the tokenized txt to normal txt
    return normal_txt
```

```python
mer_df['cln_plot'] = mer_df['plot'].apply(clean_n_normal)
```

```python
# vectorizer and Modeling
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
mlb = MultiLabelBinarizer()
```

```python
target_labels = mlb.fit_transform(mer_df['Genres'])
```

```python
target_labels
```

```python
bin_df = pd.DataFrame(target_labels,columns=mlb.classes_)
bin_df
```

```python
train_x,test_x,train_y,test_y=train_test_split(mer_df['cln_plot'],target_labels,test_size=0.2,random_state=42)
```

```python
tfidf_vec=TfidfVectorizer()
```

```python
train_x=tfidf_vec.fit_transform(train_x)
test_x=tfidf_vec.transform(test_x)
```

```python
#model
```

```python
model=LogisticRegression(max_iter=100,verbose=1)
classifier=OneVsRestClassifier(model)
```

```python
classifier.fit(train_x,train_y)
```

```python
y_pred = classifier.predict(test_x)
y_pred_label=mlb.inverse_transform(y_pred)
y_test_label=mlb.inverse_transform(test_y)
```

```python
y_pred
```

```python
accur = accuracy_score(test_y,y_pred)
```
