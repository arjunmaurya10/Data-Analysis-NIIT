```python
doc = ["This movie is very scary and long",
       "This movie is not scary and is slow",
       "This movie is spooky and good"]
```

```python
def bow_rep(document):
    bow_dict={} #blank dict for storng wrd occurance
    
    for doc in document:
        words=doc.lower().split() #lowercase and splitting words
        for word in words:
            if word in bow_dict:
                bow_dict[word]+=1 #incrementing word count
            else:
                bow_dict[word]=1
    return bow_dict  
```

```python
bow=bow_rep(doc)
print(bow)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
bow=vectorizer.fit_transform(doc)
b_array=bow.toarray()

print(b_array)
```

```python
vocab = vectorizer.get_feature_names_out()
print(vocab)
```

```python
import spacy
```

```python
from spacy.matcher import Matcher
```

```python
nlp=spacy.load("en_core_web_lg")
```

```python
matcher = Matcher(nlp.vocab)
```

```python
pattern=[{'LOWER':'hey'},{"LOWER":"siri"}]
matcher.add("heysiri",[pattern])
```

```python
text ="Hey, Siri! Hey Siri!"
```

```python
doc=nlp(text)
```

```python
matches=matcher(doc)
```

```python
for match_id,start,end in matches:
    span=doc[start:end]
    print(span.text)
```

```python
pattern1=[{'LOWER':'hey'},{"IS_PUNCT":True},{"LOWER":"siri"}]
matcher1.add("heysiri",[pattern1])
```

```python
matcher1 = Matcher(nlp.vocab)
```

```python
text1 ="Hey, Siri! Hey Siri!"
```

```python
doc1=nlp(text1)
```

```python
matches=matcher1(doc1)
```

```python
for match_id,start,end in matches:
    span=doc1[start:end]
    print(span.text)
```

```python
from spacy.matcher import PhraseMatcher
```

```python
matcher = PhraseMatcher(nlp.vocab,attr='LOWER')
```

```python
phrases = ["sweet oranges","ROTTEN mangoes"]
```

```python
phr_tk=[nlp(text.lower()) for text in phrases]
```

```python
matcher.add("phrases",None,*phr_tk)
```

```python
text = "Do not put rotten mangoes and sweet oranges together"
```

```python
doc=nlp(text)
```

```python
matches=matcher(doc)
```

```python
for match_id,start,end in matches:
    span=doc[start:end]
    print(span.text)
```
