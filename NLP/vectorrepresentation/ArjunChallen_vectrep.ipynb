```python
def bow_rep(document):
    bow_dict={} #blank dict for storng wrd occurance
    
    for d1 in document:
        words=d1.lower().split() #lowercase and splitting words
        for word in words:
            if word in bow_dict:
                bow_dict[word]+=1 #incrementing word count
            else:
                bow_dict[word]=1
    return bow_dict  
```

```python
bow=bow_rep(d1)
print(bow)
```

```python
from spacy.matcher import Matcher
```

```python
matcher = Matcher(nlp.vocab)
```

```python
pattern=[{'LOWER':'battle'},{"IS_PUNCT":True},{"LOWER":"field"}]
matcher.add("battlefield",[pattern])
```

```python
t2="We are met on a great battlefield "
```

```python
doc1=nlp(t2)
```

```python
matches=matcher(doc1)
```

```python
for match_id,start,end in matches:
    span=doc1[start:end]
    print(span.text)
```

```python
from spacy.matcher import PhraseMatcher
```

```python
matcher = PhraseMatcher(nlp.vocab,attr='LOWER')
```

```python
phrases = ["long endure","resting place"]
```

```python
phr_tk=[nlp(text.lower()) for text in phrases]
```

```python
matcher.add("phrases",None,*phr_tk)
```

```python
txt2 = "Four score and seven years ago our fathers brought forth upon this continent a new nation conceived in Liberty and dedicated to the proposition that all men are created equal Now we are engaged in a great civil war testing whether that nation or any nation so conceived and so dedicated can long endure We are met on a great battle field of that war We have come to dedicate a portion of that field as a final resting place for those who here gave their lives that that nation might live It is altogether fitting and proper that we should do this But in a larger sense we can not dedicate we can not consecrate we can not hallow this ground"
```

```python
doc=nlp(txt2)
```

```python
matches=matcher(doc)
```

```python
for match_id,start,end in matches:
    span=doc[start:end]
    print(span.text)
```
