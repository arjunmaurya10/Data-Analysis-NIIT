```python
import pandas as pd
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
dfhr = open('DS3_C2_S2_48hrs_Data_Practice.txt')
```

```python
txt = dfhr.read()
```

```python
txt
```

```python
# white spaces
txt=re.sub(r'\s+',' ',txt)
txt
```

```python
txt=re.sub(r'[^\w\s]','',txt)
txt
```

```python
#t1
```

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
#tokenization of text using spacy
doc = nlp(txt)
```

```python
for token in doc:
    print(token.text)
```

```python
nlp.disable_pipes('tagger')
```

```python
nlp.disable_pipes('lemmatizer')
```

```python
nlp.disable_pipes('ner')
```

```python
nlp.disable_pipes('parser')
```

```python
#t2
```

```python
from collections import Counter
```

```python
freq=Counter(token.text for token in doc)
freq
```

```python
freq.most_common()[:10]
```

```python
freq.get('prisoner')
```

```python
freq2=Counter(token.text for token in doc[:1500])
freq2
```

```python
freq2.get('prisoner')
```

```python
freq2.get('prisoners')
```
